##now scrapping

```python
job_role=[]
Company=[]
Salary =[]
Location=[]

base_url = "https://internshala.com"
url = base_url + "/jobs/keywords-job/"

```

```python
while url:
    print(f"Scraping: {url}")
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    job_infoo = soup.find_all('div', class_='container-fluid individual_internship view_detail_button visibilityTrackerItem')

    #jobrole
    for job in job_infoo:
        role_tag = job.find('a', class_='job-title-href')
        role = role_tag.text.strip() if role_tag else "N/A"
        job_role.append(role)

    #company
    for job in job_infoo:
        company_tag = job.find('p', class_='company-name')
        company = company_tag.text.strip() if company_tag else "N/A"
        Company.append(company)


    #Salary
    for job in job_infoo:
        salary_tag = job.find('span', class_='desktop')
        salary = salary_tag.text.strip() if salary_tag else "N/A"
        Salary.append(salary)

    #location
    for job in job_infoo:
        location_tag = job.find('p', class_='row-1-item locations')
        location = location_tag.text.strip() if location_tag else "N/A"
        Location.append(location)


            # Find next page
    next_btn = soup.find('a', class_='next_page')
    if next_btn and next_btn.get('href'):
        next_page = next_btn.get('href')
        url = base_url + next_page
    else:
        break

```

```python
print(len(job_role), len(Company), len(Location),len(Salary),len(Location))
```

#converting into dataframes

```python
df= pd.DataFrame({"JOB Role": job_role, "Company":Company ,"Salary":Salary,"Location":Location})
df
```
